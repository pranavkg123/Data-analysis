```python
# distribution of data
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#HISTOGRAM
```

```python
# Distribution plot for city_mpg
sns.histplot(df['city_mpg'],color='skyblue')
plt.title('Distribution of City MPG')
plt.show()
```

```python
# Distribution plot for cylinders
sns.histplot(df['cylinders'], kde=True, color='lightgreen')
plt.title('Distribution of Cylinders')
plt.show()
```

```python
# Distribution plot for displacement
sns.histplot(df['displacement'], kde=True, color='salmon')
plt.title('Distribution of Displacement')
plt.show()
```

```python
# Boxplot for city_mpg
sns.boxplot(data=df, x='city_mpg', color='skyblue')
plt.title('Boxplot of City MPG', fontsize=14)
plt.xlabel('City MPG', fontsize=12)
plt.show()

```

```python
# Boxplot for highWAY_MPG
sns.boxplot(data=df, x='highway_mpg', color='skyblue')
plt.title('Boxplot of HIGHWAY MPG', fontsize=14)
plt.xlabel(' MPG', fontsize=12)
plt.show()
```

```python
# Distribution of car classes
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='class', order=df['class'].value_counts().index)
plt.title('Distribution of Car Classes')
plt.xticks(rotation=45)
plt.show()
```

```python
#count plot
# Fuel type distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='fuel_type', order=df['fuel_type'].value_counts().index)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
# Correlation heatmap for numeric features
import numpy as np
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of numeric Features')
plt.show()
```

```python
# Line Plot
sns.lineplot(x='cylinders', y='highway_mpg', data=df)
plt.title('Highway MPG vs Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel(' highway MPG')
plt.show()
```

```python
# Line Plot
sns.lineplot(x='cylinders', y='city_mpg', data=df)
plt.title('City MPG vs Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('City MPG')
plt.show()
```

```python
sns.lineplot(x='displacement', y='highway_mpg', data=df)
plt.title('highway MPG vs displacement')
plt.xlabel('displacement')
plt.ylabel('highway MPG')
plt.show()
```

```python
sns.lineplot(x='displacement', y='city_mpg', data=df)
plt.title('city MPG vs displacement')
plt.xlabel('displacement')
plt.ylabel('city MPG')
plt.show()
```

```python

# Bar Plot for city_mpg and car class
plt.figure(figsize=(8, 4))
sns.barplot(x='class', y='city_mpg', data=df)
plt.title('Bar Plot')
plt.xticks(rotation=45)

plt.show()
```

```python
# Bar Plot for city_mpg and car class
plt.figure(figsize=(8, 4))
sns.barplot(x='class', y='highway_mpg', data=df)
plt.title('Bar Plot')
plt.xticks(rotation=45)

plt.show()
```
