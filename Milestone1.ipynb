#                                            COFFEE DATA  ANALYSIS

Coffee is one of the world's most popular beverages, along with water and tea. It's made from the roasted and ground seeds of the tropical evergreen coffee plant, which originated in Africa. The two most common types of coffee beans are C. arabica and C. robusta






Coffee quality is determined by its aroma, taste, and physical characteristics. The quality of coffee is influenced by many factors, including the growing conditions, the processing, and the roasting.
Factors that affect coffee quality

Growing conditions: Altitude, shade, and post-harvest processing can affect the accumulation of non-volatile compounds in coffee beans.


Moisture content: Beans that are too wet can mold, while beans that are too dry lose flavor.


Roasting: The temperature and duration of roasting can affect the aroma and flavor of the coffee. Roasting for too long can destroy the volatile compounds that give coffee its flavor.


Characteristics that determine coffee quality:
Physical characteristics: The length, width, shape, color, and weight of the coffee beans


Organoleptic characteristics: The aroma, flavor, sweetness, acidity, aftertaste, bitterness, body, and balance of the coffee

Biochemical composition: The levels of caffeine, chlorogenic acid (CGA), trigonelline, lipids, and sucrose in the coffee beans


Methods for assessing coffee quality


Analytical chemistry: Using chemical analysis and chromatographic techniques to measure the biochemical composition of the coffee

Sensory evaluation: Using cuppers to determine the cup quality of the coffee
